# Backprop algorithm

Note, this is w.r.t. sum of squared error (SSE).

---

* Pick your MLP architecture
  
  * number of hidden layers, activation functions, etc.

* Training data
   
  * $T = \{ (\textbf{x}(k),\textbf{d}(k)) \}_{k=1}^{K}$
  
  * note: $\textbf{x}$ and $\textbf{d}$ are vectors and $(k)$ is one instance

* Init weights

  * e.g., random values

* Repeat
  
  * Do, for $k = 1,\ldots,K$ (one epoch)
  
     * Forward computation (layer-by-layer)
     
       * Do, for $h = 1, 2, \ldots, $ to $H$
      
          * compute $v_j^h(k) = \left( \textbf{w}_j^h(k) \right)^t \textbf{y}^{h-1}(k)$
            
            * $j$th dot product at layer $h$
            
            * note: $\textbf{y}^0(k) = \textbf{x}(k)$
        
          * compute $\textbf{y}^h_j(k) = \varphi_j\left( v_j^h(k) \right)$
          
            * $j$th activation at layer $h$
            
       * End for (forward pass)
                
     * Error
       
       * compute $e_j^H(k) = d_j(k) - y_j^H(k)$

     * Backward       
     
       * Do, for $h = H, H-1, \ldots,$ to $1$
     
         * compute $\delta_j^h(k) = \left\{ \begin{array}{cc} 
                e^H_j(k) \varphi^{'}(v_j^H(k)) & \text{if at output layer } (h == H) \\
                \varphi^{'}(v_j^h(k)) \sum_{l=1}^{L}{ \delta_l^{h+1}(k)w_{lj}^{h+1}(k) } & \text{else} \\
                \end{array} \right.$
                
         * weight update 
     
            * $w_{ji}^h(k+1) = w_{ji}^h(k) + \alpha \Big[ \delta_j^h(k) y_i^{h-1}(k) \Big]$       
       
     * End for (backward pass) 
            
  * End for (epoch)
  
  * Shuffle the data
  
* Until DONE
